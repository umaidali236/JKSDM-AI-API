```python
import pandas as pd
import random
from tqdm import tqdm
```

```python
self_learning_videos = pd.read_csv('./db/self_learning_courses.csv')
```

```python
self_learning_videos['BASIC_ADVANCED'] = ''
```

```python
sectors = list(self_learning_videos['sector'].unique())
```

```python
map_course_to_basic_advanced = dict()
for sector in tqdm(sectors):
    map_course_to_basic_advanced[sector] =dict()
    print(sector)
    #get courses from a sector
    self_learning_videos_in_a_sector = self_learning_videos[self_learning_videos['sector']==sector]
    courses_in_a_sector = list(self_learning_videos_in_a_sector['name of course'].unique())
    total_courses_in_a_sector= len(courses_in_a_sector)
    if total_courses_in_a_sector < 10:
        some_random_courses_in_a_sector = random.sample(courses_in_a_sector,total_courses_in_a_sector)
    else:
        some_random_courses_in_a_sector = random.sample(courses_in_a_sector,10)
    len_= len(some_random_courses_in_a_sector) #len of sampled courses
    
    for course in courses_in_a_sector:
        
        for c in some_random_courses_in_a_sector[0:len_//2]:
            map_course_to_basic_advanced[sector][c] = 'BASIC'
        for c in some_random_courses_in_a_sector[len_//2:]:
            map_course_to_basic_advanced[sector][c] = 'ADVANCED'    
```

```python
from transformers import BertTokenizer, BertModel
import torch
```

```python

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
model.eval()




def get_bert_embedding(text):
    # Tokenize input text and convert to PyTorch tensors
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Get the embeddings for the [CLS] token
    embeddings = outputs.last_hidden_state[:, 0, :].squeeze()
    return embeddings
```

```python
self_learning_courses = dict()
self_learning_courses_embedding_map = dict()
```

```python
self_learning_videos = pd.read_csv('./db/self_learning_courses_basic_advanced.csv')
```

```python
sectors = list(self_learning_videos['sector'].unique())
for sector in tqdm(sectors):
    courses = list(self_learning_videos['name of course'].unique())
    for c in courses:
        if not str(c)=="nan":

            self_learning_courses_embedding_map[c] = get_bert_embedding(c)
```

```python
embeddings_path=list()
for c in tqdm(list(self_learning_courses_embedding_map.keys())):
    print(c)
    filename = str(c).replace("/","").replace("|","")  
    embeddings_path.append([c,filename+".pt"])
    torch.save(self_learning_courses_embedding_map[c], './db/course_embeddings/'+str(filename)+".pt")
```

```python
pd.DataFrame(embeddings_path,columns=['name of course','embedding_path']).to_csv('./db/embeddings_path.csv')
```
